```python
def Chi_2(d1, dist, titulo):
    dist = getattr(st, dist)
    J = len(d1)
    
    #Calcular la frecuecia 
    hist, bin_edges = np.histogram(d1, bins=(J))
    bins = bin_edges[1]-bin_edges[0]
    freq = (hist/(bins*sum(hist)))
    
    x = np.linspace(min(d1),max(d1), J)
    
    param = st.beta.fit(d1)
    pi = st.gengamma.pdf(x, *param[:-2], loc=param[-2], scale=param[-1])
    
    Y_np = np.array(d1)
    E_i = Y_np*pi
    chi_est = st.chisquare(freq,E_i)
    print('Valor de chi cuadrado librería = ', chi_est)

    # Cálculo de Grados de libertad del estadístico
    p = 2 
    m = J-p-1 # grados de libertad
    Chi_est = st.chi2.ppf(q = 0.95,df=m)
    print('Estadístico de chi_cuadrado = ',Chi_est)
```

```python
def kde_statsmodels_u(x, x_grid, bandwidth, **kwargs):
    """Univariate Kernel Density Estimation with Statsmodels"""
    kde = KDEUnivariate(x)
    kde.fit(bw = bandwidth, **kwargs)
    return lambda x: kde.evaluate(x)
```

```python
def Chi_2(d2, dist, titulo):
    dist = getattr(st, dist)
    J = len(d2)
    
    #Calcular la frecuecia 
    hist, bin_edges = np.histogram(d2, bins=(J))
    bins = bin_edges[1]-bin_edges[0]
    freq = (hist/(bins*sum(hist)))
    
    x = np.linspace(min(d2),max(d2), J)
    
    param = st.genpareto.fit(d2)
    pi = st.genpareto.pdf(x, *param[:-2], loc=param[-2], scale=param[-1])
    
    Y_np = np.array(d2)
    E_i = Y_np*pi
    chi_est = st.chisquare(freq,E_i)
    print('Valor de chi cuadrado librería = ', chi_est)

    # Cálculo de Grados de libertad del estadístico
    p = 2 
    m = J-p-1 # grados de libertad
    Chi_est = st.chi2.ppf(q = 0.95,df=m)
    print('Estadístico de chi_cuadrado = ',Chi_est)
```
